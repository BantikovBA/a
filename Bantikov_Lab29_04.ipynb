Какую массу имеет метан при постоянном 

T=273.15  P=1.0133*10^5 V=64 m^3

```python
import numpy as np
import CoolProp.CoolProp as ct
import matplotlib.pyplot as plt
p=101325 #Давление 1 атм
V=64     #Объём
R=8.3145 #Универсальная газовая постоянная
M=0.016  #Молярная масса метана
T=np.linspace(100,150,1000) #Задание температуры
m1=(p*V*M)/(R*T)   #МРешение 1
D=ct.PropsSI("D","T",T,"P",p,"CH4") #Плотность по CollProp
m2=D*V  #Решение 2
plt.plot(T,m1,"b")
plt.plot(T,m2,"r")
plt.xlabel("T")
plt.ylabel("m")
plt.title("Зависимость m от T при постоянном давлении")
plt.legend(['m школьная','m coolprop'])
plt.scatter(111.65,ct.PropsSI("D","T",111.65,"P",p,"CH4"),color="green")
```
